## CIFAR-10 Data Loading and Preprocessing

###  Data Loading

```python
# Load the raw CIFAR-10 data.
cifar10_dir = 'datasets/cifar-10-batches-py/'
X_train, y_train,  X_test, y_test = load_CIFAR10(cifar10_dir)

print ('X_train shape:  ',X_train.shape)
print ('y_train shape:  ',y_train.shape)
print ('X_test shape:  ',X_test.shape)
print ('y_test shape:  ',y_test.shape)
```

```python
# visualize 
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
num_classes = len(classes)
sample_each_class = 5

for y , cls in enumerate(classes):
    idxs = np.flatnonzero(y_train == y)
    idxs = np.random.choice(idxs, sample_each_class, replace=False)
    for  i, idx in enumerate(idxs):
        plt_idx = i * num_classes + y + 1
        plt.subplot(sample_each_class, num_classes, plt_idx)
        plt.imshow(X_train[idx].astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(cls)
            
plt.show()
```

```python
# Split the data into train, val, and test sets.
# 创建一个 小的 development set 作为 训练的子集，有助于快速 运行 code

num_train = 49000
num_val = 1000
num_test = 10000
num_dev = 500

# validation set
mask = range(num_train, num_train + num_val) # 49000 - 50000
X_val = X_train[mask]
y_val = y_train[mask]

# training set 
mask = range(num_train)   # 0 - 49000
X_train = X_train[mask]
y_train = y_train[mask]

# development set
mask = np.random.choice(num_train, num_dev, replace=False)  # 从 49000 中随机抽 500
X_dev = X_train[mask]
y_dev = y_train[mask]

# test set    # 0 - 10000
mask = range(num_test)
X_test = X_test[mask]
y_test = y_test[mask]
```

```python
# cheak 
print ('X_train shape:  ',X_train.shape)
print ('y_train shape:  ',y_train.shape)
print ('X_val shape:    ',X_val.shape)
print ('y_val shape:    ',y_val.shape)
print ('X_test shape:   ',X_test.shape)
print ('y_test shape:   ',y_test.shape)
print ('X_dev shape:    ',X_dev.shape)
print ('y_dev shape:    ',y_dev.shape)
```

###  proprocessing : reshape

```python
# reshape the image data into rows
X_train = np.reshape(X_train, (X_train.shape[0], -1)) 
X_val = np.reshape(X_val, (X_val.shape[0], -1)) 
X_test = np.reshape(X_test, (X_test.shape[0], -1)) 
X_dev = np.reshape(X_dev, (X_dev.shape[0], -1)) 
```

```python
# cheak 
print ('X_train shape:  ',X_train.shape)
print ('X_val shape:    ',X_val.shape)
print ('X_test shape:   ',X_test.shape)
print ('X_dev shape:    ',X_dev.shape)
```

- proprocessing : subtract the mean image 减去 图像均值

```python
print("Before : \n X_train[0]", X_train[0])
plt.figure(figsize=(4, 4))
plt.imshow(X_train[0].reshape(32,32,3).astype('uint8'))
plt.show()
```

```python
#  1、计算 train data 的均值图像
mean_img = np.mean(X_train, axis=0)


print(mean_img[:10])
plt.figure(figsize=(4, 4))
plt.imshow(mean_img.reshape(32,32,3).astype('uint8'))
plt.show()
```

```python
#  2、 train 和 test data 减去 均值图像
X_train -= mean_img
X_val -= mean_img
X_test -= mean_img
X_dev -= mean_img
```

```python
print("After : \n X_train[0]", X_train[0])
plt.figure(figsize=(4, 4))
plt.imshow(X_train[0].reshape(32,32,3).astype('uint8'))
plt.show()
```

```python
# 3、每个 image 增加 1维的 bias
X_train = np.hstack([X_train, np.ones((X_train.shape[0], 1))])  # n x 1dim
X_val = np.hstack([X_val, np.ones((X_val.shape[0], 1))]) 
X_test = np.hstack([X_test, np.ones((X_test.shape[0], 1))]) 
X_dev = np.hstack([X_dev, np.ones((X_dev.shape[0], 1))]) 
```

```python
# cheak 
print ('X_train shape:  ',X_train.shape)
print ('X_val shape:    ',X_val.shape)
print ('X_test shape:   ',X_test.shape)
print ('X_dev shape:    ',X_dev.shape)
```

```python
# 梯度检查：分析梯度的结果 与 数值梯度结果 进行比较
from gradient_check import grad_check_sparse

# 1、 损失函数 不加入 正则化损失时，进行 梯度检查
print ("without  regularation")
loss , grad = svm_loss_naive(W, X_dev, y_dev, 0.0)
f = lambda w: svm_loss_naive(w, X_dev, y_dev, 0.0)[0]
grad_numerical = grad_check_sparse(f, W, grad)


# 2、损失函数 加入 正则化损失时，进行 梯度检查
print ("\n with  regularation ")
loss , grad = svm_loss_naive(W, X_dev, y_dev, 5e1)
f = lambda w: svm_loss_naive(w, X_dev, y_dev, 5e1)[0]
grad_numerical = grad_check_sparse(f, W, grad)
```

#### 比较 不同方法计算loss 的运行时间，以及loss 的差距

```python
# 用 naive 方法计算的 loss 时的运行时间
start = time.time()
loss_naive, _ = svm_loss_naive(W, X_dev, y_dev, 0.000005)
end = time.time()
print ("Naive loss : %e computer in %fs" %(loss_naive, end - start))

# 用 vectorized 方法计算的 loss 时的运行时间
start = time.time()
loss_vectorized, _ = svm_loss_vectorized(W, X_dev, y_dev, 0.000005)
end = time.time()
print ("Vectorized loss : %e computer in %fs" %(loss_vectorized, end - start))


# 比较 两个loss 的差距
different = loss_naive - loss_vectorized  # loss 是 single number
print ('different: %f' % (different))

"""
# train set 的结果
Naive loss : 8.805583e+00 computer in 38.725507s
Vectorized loss : 8.805583e+00 computer in 14.546225s   -- good
different: -0.000000
"""
```

#### 比较 不同方法 计算 gradient 的运行时间，以及gradient 的差距

```python
# 用 naive 方法计算的 gradient 时的运行时间
start = time.time()
_, grad_naive = svm_loss_naive(W, X_dev, y_dev, 0.000005)
end = time.time()
print ("Naive gradient : computer in %fs" %( end - start))

# 用 vectorized 方法计算的 gradient 时的运行时间
start = time.time()
_, grad_vectorized = svm_loss_vectorized(W, X_dev, y_dev, 0.000005)
end = time.time()
print ("Vectorized gradient :  computer in %fs" %( end - start))


# 比较 两个gradient 的差距
different = np.linalg.norm(grad_naive - grad_vectorized, ord='fro')  # 梯度是 matrix
print ('different: %f' % (different))
```

## training -- >SGD -- Stochastic Gradient Descent

    class LinearSVM(LinearClassifier):
      """ A subclass that uses the Multiclass SVM loss function """

      def loss(self, X_batch, y_batch, reg):
        return svm_loss_vectorized(self.W, X_batch, y_batch, reg)


```python
from classifiers.linear_classifier import LinearSVM

# 使用 train set 正式 训练
svm = LinearSVM()
start = time.time()
loss_hist = svm.train(X_train, y_train, learning_rate=1e-7, reg=2.5e4, num_iters=1500,verbose=True)

stop = time.time()
print ('That took %fs' %(stop - start))
```

```python
# 画图 损失 随迭代的变换
plt.plot(loss_hist)
plt.xlabel('Interation number')
plt.ylabel('loss value')
plt.show
```

```python
# 查看 预测结果 的准确率
y_train_pred = svm.predict(X_train)
print ('training accuracy: %f' %(np.mean(y_train == y_train_pred),))

y_val_pred = svm.predict(X_val)
print ('validation accuracy: %f' %(np.mean(y_val == y_val_pred),))
```

## 调整超参数 --> 交叉验证  on  val set 
    regularization strength
    learning rate

```python
# 设置 需要调整的参数 及其 范围
learning_rate = [1e-7, 5e-5, 4e-3]
regularization_strength = [2.5e4, 5e4, 8e4]
```

```python
# 交叉 验证

# 通过在 val set 上调整 选择最好的超参
# 每个超参数组合，
# 在train set 上训练 分类器
# 在 train set 和 val set 上 计算 准确率


results = {}  #(learning_rate, regularization_strength)
best_val = -1 # 最高的 验证准确率 
best_svm = None # 获得最高 验证准确率的 LinearSVM object

num_iters = 10

for lr in learning_rate:
    for reg in regularization_strength:
        print ("\n lr:reg   ",lr,reg)
        
        # 1、创建 分类器
        svm = LinearSVM()
        
        # 2、训练 分类器
        loss = svm.train(X_train, y_train, learning_rate=lr, reg=reg, num_iters=num_iters,verbose=False)

        # 3、计算 准确率
        y_train_pred = svm.predict(X_train)
        train_acc = np.mean(y_train == y_train_pred)

        y_val_pred = svm.predict(X_val)
        val_acc = np.mean(y_val == y_val_pred)
        
        # 4、存储 结果    
        results[(lr, reg)] = (train_acc, val_acc) 
        
        if best_val < val_acc:
            best_val = val_acc
            best_svm = svm
            
        
```

```python
# print  cross-validation results
for lr, reg in sorted(results):
    train_acc, val_acc = results[(lr, reg)]
    print ('lr:%e  reg:%e  --> train accuracy:%f ; val accuracy:%f ' % (
          lr, reg, train_acc, val_acc) )
print ('Best validation accuracy achieved during cross-validation: %f' % best_val)
```

#### 可视化  cross-validation results

```python

import math
x_scatter = [math.log10(x[0]) for x in results]
y_scatter = [math.log10(x[1]) for x in results]

# plot training accuracy
market_size = 100
colors = [results[x][0] for x in results]
plt.subplot(2, 1, 1)
plt.scatter(x_scatter, y_scatter, market_size, c=colors)
plt.colorbar()
plt.xlabel('log learning rate')
plt.ylabel('log regularzation strength')
plt.title('CIFAR-10 training accuracy')

# plot validation accuracy
market_size = 100
colors = [results[x][1] for x in results]
plt.subplot(2, 1, 2)
plt.scatter(x_scatter, y_scatter, market_size, c=colors)
plt.colorbar()
plt.xlabel('log learning rate')
plt.ylabel('log regularzation strength')
plt.title('CIFAR-10 validation accuracy')

plt.show()
```

####  可视化 每个类别 W 

```python

# W 是 根据 选择的lr reg 学习到的

w = best_svm.W[:-1, :] # 排除 bias
w = w.reshape(32, 32, 3, 10)
w_min, w_max = np.min(w), np.max(w)

classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

for i in range(10):
    plt.subplot(2, 5, i + 1)
    
    # rescale the weight to be between 0 and 255
    wimg = 255.0 * (w[:, :, :, i].squeeze() - w_min) / (w_max - w_min)
    
    plt.imshow(wimg.astype('uint8'))
    plt.axis('off')
    plt.title(classes[i])
    
```
